```python
def is_product_odd(lst):
    product = 1
    for num in lst:
        product *= num
    return product % 2 != 0
```

```python
def count_vowels(s):
    vowels = 'aeiouAEIOU'
    return sum(1 for char in s if char in vowels)
```

```python
def caesar_cipher(text, key):
    result = ''
    for char in text:
        if char.isalpha():
            shift = key if char.islower() else key
            start = ord('a') if char.islower() else ord('A')
            result += chr((ord(char) - start + shift) % 26 + start)
        else:
            result += char
    return result

key = int(input("Enter key: "))
text = input("Enter plaintext: ")
print("Encrypted:", caesar_cipher(text, key))
```

```python
def brute_force_caesar(ciphertext):
    for key in range(26):
        decrypted = ''
        for char in ciphertext:
            if char.isalpha():
                start = ord('A') if char.isupper() else ord('a')
                decrypted += chr((ord(char) - start - key) % 26 + start)
            else:
                decrypted += char
        print(f'Key {key}: {decrypted}')

ciphertext_list = [
    "R UXEN VH TRCCH",
    "FR DBMMR EHOXL FX",
    "CXPNCQNA FN'AN BX QJYYH",
    "OBR OZKOMG QOFSTFSS",
    "PDKQCD IU DAWZ DWO OQOLEYEKJO",
    "FTMF U WQQB GZPQD YK TMF",
    "AR ITMF YUSTF TMBBQZ",
    "DA D NCMVIF OJ OCZ NDUZ JA V MVO",
    "ZFBI. J'N QSFUUZ TVSF NZ DBU XPVME FBU NF"
]

for ciphertext in ciphertext_list:
    brute_force_caesar(ciphertext)

```
